```python
import pandas as pd
```

```python
AirQuality = pd.read_csv("LA_daily_air_quality.csv")
```

```python
AirQuality.head()
```

```python
AirQuality.info()
```

```python
AirQuality['ID'] = (AirQuality['latitude'].astype(str)+ AirQuality['longitude'].astype(str)+ AirQuality['date'].astype(str))

AirQuality.head()
                                    
```

```python
DailyClimate = pd.read_csv('LA_daily_climate.csv')
DailyClimate.head()
```

```python
DailyClimate.info()
```

```python
DailyClimate['ID'] = (DailyClimate['latitude'].astype(str)+ DailyClimate['longitude'].astype(str)+ DailyClimate['date'].astype(str))

DailyClimate.head()
```

```python
DailyClimate_AirQuality = DailyClimate.merge(AirQuality, how = 'inner', on = 'ID')
```

```python
DailyClimate_AirQuality.head()
```

```python
DailyClimate_AirQuality.columns
```

```python
DailyClimate_AirQuality = DailyClimate_AirQuality.drop(['date_y', 'latitude_y', 'longitude_y'], axis=1)
DailyClimate_AirQuality.head()
```

```python
DailyClimate_AirQuality.describe()
```

```python
Country_City_group_wind_speed_ozone = DailyClimate_AirQuality.groupby(['country', 'city'])[['wind_speed_10m_max', 'ozone']].mean()
Country_City_group_wind_speed_ozone.head()
```

```python
Country_City_group_wind_speed_ozone['ozone'].corr(Country_City_group_wind_speed_ozone['wind_speed_10m_max'])
```

Dakle korelacija izmedju prosecne brzine vetra i prosecnog nivoa ozona nije znacajna.


```python
Country_City_group_wind_speed_carbon_monoxide = DailyClimate_AirQuality.groupby(['country', 'city'])[['wind_speed_10m_max', 'carbon_monoxide']].mean()
Country_City_group_wind_speed_carbon_monoxide.head()
```

```python
Country_City_group_wind_speed_carbon_monoxide['carbon_monoxide'].corr(Country_City_group_wind_speed_carbon_monoxide['wind_speed_10m_max'])
```

ne postoji znacajna korelacija izmedju prosecnog nivoa karbon monoksida i prosecne maksimalne brzine vetra u glavnim gradovima drzava


```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(Country_City_group_wind_speed_carbon_monoxide['carbon_monoxide'],Country_City_group_wind_speed_carbon_monoxide['wind_speed_10m_max'])
plt.ylabel('Average carbon monoxide')
plt.xlabel('Average max wind speed')
```

```python
Country_City_group_temperature_2m_mean_carbon_monoxide = DailyClimate_AirQuality.groupby(['country', 'city'])[['temperature_2m_mean', 'carbon_monoxide']].mean()
Country_City_group_temperature_2m_mean_carbon_monoxide.head()
```

```python

plt.scatter(Country_City_group_temperature_2m_mean_carbon_monoxide['carbon_monoxide'], 
            Country_City_group_temperature_2m_mean_carbon_monoxide['temperature_2m_mean'])

# Adding labels and title
plt.xlabel('Average Carbon Monoxide')
plt.ylabel('Average Temperature (2m mean)')
plt.title('Scatter Plot of Carbon Monoxide vs Temperature')
```

vizuelno nema korelacije ni izmedju proseka prosecnih temperatura tokom godine i nivoa karbon monoxida


```python
Country_City_latitude_temp_max = DailyClimate_AirQuality.groupby(['country', 'city'])[['latitude_x', 'temperature_2m_max']].mean()
Country_City_latitude_temp_max.head()
```

```python
Country_City_latitude_temp_max['temperature_2m_max'].corr(Country_City_latitude_temp_max['latitude_x'])
```

Postoji slaba korelacija izmedju geografske sirine i prosecno izmerene maksimalne temperature.

```python
plt.scatter(Country_City_latitude_temp_max['temperature_2m_max'], Country_City_latitude_temp_max['latitude_x'])
```
