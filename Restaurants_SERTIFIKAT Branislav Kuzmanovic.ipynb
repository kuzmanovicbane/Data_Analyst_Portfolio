```python
import pandas as pd
```

```python

restaurant_data = pd.read_excel('restaurants.xlsx', sheet_name = 'restaurant_data')
restaurant_data
```

```python
review_data = pd.read_excel('restaurants.xlsx', sheet_name = 'review_data')
review_data
```

```python
restaurant_data.info()
```

```python
restaurant_data.isna().sum()
```

```python
mode_Smoking_Allowed = restaurant_data['Smoking_Allowed'].mode()[0]
mode_Smoking_Allowed
```

```python
restaurant_data['Smoking_Allowed'] = restaurant_data['Smoking_Allowed'].fillna(mode_Smoking_Allowed)
restaurant_data.isna().sum()
```

```python
mode_Price = restaurant_data['Price'].mode()[0]
mode_Price
```

```python
restaurant_data['Price'] = restaurant_data['Price'].fillna(mode_Price)
restaurant_data.isna().sum()
```

```python
restaurant_data['Name'].duplicated(keep = False)
```

```python
columns = ['Name','City']
```

```python
restaurant_data[restaurant_data[columns].duplicated(keep = False)].sort_values('Name')
```

```python
columns = ['Name', 'City']
restaurant_data = restaurant_data.drop_duplicates(subset=columns, keep='first')
restaurant_data
```

```python
review_data.info()
```

```python
review_data.isna().sum()
```

```python
average_Food_Rating = review_data['Food_Rating'].mean()
average_Food_Rating
```

```python
review_data['Food_Rating'] = review_data['Food_Rating'].fillna(average_Food_Rating)
review_data.isna().sum()
```

```python
average_Service_Rating = review_data['Service_Rating'].mean()
average_Service_Rating
```

```python
review_data['Service_Rating'] = review_data['Service_Rating'].fillna(average_Service_Rating)
review_data.isna().sum()
```

```python
restaurant_data.info()
```

```python
review_data.info()
```

```python
restaurant_data.to_csv('restaurant_data_za_power_bi_1.csv',index = False)
```

```python
review_data.to_csv('review_data_za_power_bi_1.csv', index = False)
```

```python
review_data['Overall_Rating'].corr(review_data['Food_Rating'])
```

```python
review_data['Overall_Rating'].corr(review_data['Service_Rating'])
```

definitivno je veca korelacija izmedju Food Rating-a i Overall Rating-a nego izmedju Service Rating-a i Overall Rating-a.

```python
grouped_review_data = review_data.groupby('Restaurant_ID')['Overall_Rating'].mean()
grouped_review_data
```

```python
restaurant_with_overall_review = restaurant_data.merge(grouped_review_data, how = 'inner',on = 'Restaurant_ID')
restaurant_with_overall_review
```

```python
group_Price_AlcoholService = restaurant_with_overall_review.groupby(['Price', 'Alcohol_Service'])['Overall_Rating'].mean().reset_index()

group_Price_AlcoholService.sort_values(by='Overall_Rating', ascending=False)
```

Vidimo da restorani koji su u kategoriji najskupljih u proseku imaju najvece rejtinge. U istoj kategoriji sirina asortimana alkoholnih pica igra ulogu u rejtingu, odnosno prosecan rejting je veci sto je siri asortiman. U kategoriji srednje skupih sirina asortimana ne utice na prosecan rejting, a u kategoriji jeftinih visu prosecnu ocenu ima Full Bar, ali No Alcohol ili Wine and Beer imaju slican rejting.
